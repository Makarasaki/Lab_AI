```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.metrics
import sklearn.datasets
import sklearn.model_selection
```

```python
def number_to_name(flower_number):
    flowers_types = ['Setosa', 'Versicolor', 'Virginica']
    return flowers_types[flower_number]
```

```python
iris_dataset = sklearn.datasets.load_iris()
dimensions_df = pd.DataFrame(data = iris_dataset.data, columns = iris_dataset.feature_names)
names_df = pd.DataFrame(data = iris_dataset.target, columns = ['species'])
names_df['species'] = names_df['species'].apply(number_to_name)
dimensions_df = pd.concat([dimensions_df, names_df], axis= 1)
sns.pairplot(dimensions_df, hue= 'species')
```

```python
input = iris_dataset['data']
output = iris_dataset['target']
input_train, input_test, output_train, output_test =sklearn.model_selection.train_test_split(input, output, test_size = 0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier as knnclass

knn = knnclass(n_neighbors=4)
knn.fit(input_train, output_train)

knn_predictions = knn.predict(input_test)
knn_accuracy = sklearn.metrics.accuracy_score(output_test, knn_predictions)

print(f'KNN accuracy: {knn_accuracy}')
print(
f"""Confusion matrix of KNN: 
{sklearn.metrics.confusion_matrix(output_test, knn_predictions)}""")
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(input_train, output_train)

svc_predictions = svc.predict(input_test)
svc_accuracy = sklearn.metrics.accuracy_score(output_test, svc_predictions)

print(f'SVC accuracy: {svc_accuracy}')
print(
f"""Confusion matrix of SVC: 
{sklearn.metrics.confusion_matrix(output_test, svc_predictions)}""")
```

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression()
linear_reg.fit(input_train, output_train)
linear_reg_predictions = linear_reg.predict(input_test)


print(f'Linear Regression accuracy: {linear_reg.score(input_test, output_test)}')

print('Mean Absolute Error:', sklearn.metrics.mean_absolute_error(output_test, linear_reg_predictions))
print('Mean Squared Error:', sklearn.metrics.mean_squared_error(output_test, linear_reg_predictions))
print('Mean Root Squared Error:', np.sqrt(sklearn.metrics.mean_squared_error(output_test, linear_reg_predictions)))
```

```python
from sklearn.linear_model import LogisticRegression

logistic_reg = LogisticRegression()
logistic_reg.fit(input_train, output_train)
logistic_reg_predictions = logistic_reg.predict(input_test)


print(f'Linear Regression accuracy: {logistic_reg.score(input_test, output_test)}')

print('Mean Absolute Error:', sklearn.metrics.mean_absolute_error(output_test, logistic_reg_predictions))
print('Mean Squared Error:', sklearn.metrics.mean_squared_error(output_test, logistic_reg_predictions))
print('Mean Root Squared Error:', np.sqrt(sklearn.metrics.mean_squared_error(output_test, logistic_reg_predictions)))
```
